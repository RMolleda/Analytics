## Working on Jsons

```python
import requests
import json
import pandas as pd
```

```python
r = requests.get(url='http://data.nba.net/prod/v2/2018/teams.json')
jason_readed = r.json()
```

```python
with open("jason_NBA.json", "w+") as outfile:
    json.dump(jason_readed, outfile, indent=4)

```

```python
with open ("jason_NBA.json") as js:
    nba18 = json.load(js)

 # He dado tantas vueltas por que en un primer intento he trabajado con el json sin importarlo desde el ordenador, sino como un dataframe a partir de los datos del link.   
```
