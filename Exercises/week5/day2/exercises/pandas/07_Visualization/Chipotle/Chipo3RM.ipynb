### Step 5. Create a histogram of the top 5 items bought

```python
quest5 = chipo.groupby(["item_name"]).agg({"quantity": "sum"}).sort_values("quantity", ascending=False).head(5)
```

```python
quest5.plot(kind="bar")
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo.item_price = chipo.item_price.str.slice(1).astype(float)
chipo
```

```python
chipo.plot(x="item_price", y="quantity", kind="scatter")
```

```python
#segunda version
x2 = chipo.groupby(["order_id"]).sum()
plt.scatter(x= x2["item_price"], y=x2["quantity"], c="green")
plt.title("Esta es la versiÃ³n currada")
plt.xlabel("Precio por pedido")
plt.ylabel("Cantidad")
plt.ylim(0)
```
